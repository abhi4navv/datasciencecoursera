# Understanding Decision Tree in Depth

# Introduction

In Machine Learning we can  divided different algorithm into two main groups first is regression and second one is classification.
A Decision Tree (Classifier) is a type of supervised learning and belongs to classification model (Decision Tree can be used for regression too by using regressor class of it).


    







# Supervised Learning-

In supervised learning the models are trained using available historic data , in this the model is provided with set of  features with its values and desired ouput for those and machine tries to learn by these,  it compares ouput created by it and actual ouput and corrects itself if some error is there. By doing so it creates a pattern which is used to predict new values when a fresh set of data are provided.

So in other words we supervise the way machine'slearning by supplying historic desired output.


# Classification Model -

As name suggests a classification model tries to divide data into different classes based on certain criteria. For example, in farming we divide crops into winter(Rabi)  and Kharif crops based on weather condition suitable for them, so in other words we can say that we classify the crops based on certain conditions like temperature needed, water requirement ,time to harvest etc, and so if we have a new crop and relevent data available then we can use classification model which will test available data on the different conditions and will give us its class(Rable or Kharif).


A decision tree uses tree like structure to splits each data of data set based on different conditions and as we know for checking condition we use if else, so we can say that decision tree is a combination of several if else statements. Each branch in image below is a result of an if statement which checks the data against different attributes or conditions.




					

Root Node- The first or main root which actually has entire sample and data start getting devided from this node. This node is important as this creates base for further nodes and also helps in creating more efficient sequence.

Leaf Nodes-These are the last set of nodes or final decisive nodes and so we can say that these nodes have same or maximum similarity in predicting the final class. 



# Why we need multiple if else statements?



The main reason is behind it is that we can not decide that a given data belongs to a particular class just by checking it against one attribute or parameter on which we classify it.

For example if we want to create a model which can decide if a loan applicant in bank belongs to a good borrower category or a bad borrower category, in this case some body may think if we check a condition that if applicants income is more than certain threshold amount set by bank then bank can consider him or her as a eligible applicant, it can be appropriate decision if this is the first loan of that person or in case the their track record with previous loan is fine, but in case if the applicant has already pending loans or had very bad track record in paying EMIs , in such cases this decision will be wrong. So we will add further nodes to check previous record and other important conditions.

So each addition of a node to check different attributes makes system more accurate, and thus tree like structure is important to make our system more robust and accurate.


# How does decision tree works?

The main goal while designing a decision tree is to get best suitable sequence in which each condition check or nodes can be arranged so that we can get best accuracy out of it and also each decision splits the data in two branches so each split should be informative or in other words each split should be able to classify data with best accuracy.

# How to find best suitable sequence ?

There are two main algorithms Information Gain and Gini Index to select best suitable sequence.

Information Gain-By using information gain method we try to find information contained by each attribute and then a attribute with maximum information will be used a root node. To calculate information gain we use entropy.

Entropy is the measure of randomness or uncertainty in a random variable ,and information gain calculates the expected reduction in entropy due to sorting of attributes.


# Mathematical Explanation

 Entropy is the measures of impurity, disorder or uncertainty in a bunch of examples.

# Gini Index-

Gini index is measure of chances that a randomly selected is classified to a incorrect class. So in case of Gini index we will chose the feature with lowest Gini index as root node because it says incorrectness in classification so if incorrectness is less means system will be more efficient .

The limitation with Gini Index is that it works for 2 classes cases only means , when we have divide data in two categories only then in that case only it will work.

Formula for calculating Gini Index is-


We will use same data set which we used or information gain to calculate Gini Index too.

For example-

FP - False Positive (False here mean the incorrect values for predicting Positive values)

FN- False Negative(False here means the incorrect values for predicting Positive Values)

# Recall and Precision-
The score method is not a reliable technique to check the efficiency of a model, there is a famous example for it is the model designed to predict the chances that a terrorist will board a flight , the model predicted that 99.99% chances are there that boarded passenger will not  be a terroristor we can say there is no chance that a terrorist will board the flight but in reallity there were 19 such passengers, so we can say accuracyscore should not be a single crieteria so the other efficient methods are-

## Recall-
Recall is the number of true positive devided by addition of true positive and number of false negative. Here true positive are poositive class prediction which are correct and false negative are data which were classified  as negative but are actuaaly positive so prediction was wrong.

Recall=TP/(TP+ FN)

# Combining Precision and Recall (F1 Score)-
As know that the recall give us ability to find all relevent cases ,the precision give us the ability to find that data which were model says is relevent si actually relevent or not.

The recall and precision are inversily proportional,so we can say that if we increase recall our precision will compromise ,and maximizing one  depends on the condition that whether we want more relvence or more precision .

We have one method to find optimal blend of these two, i.e. is F1 score .
F1 score is the harmonic mean of recall and precision

F1 = 2* (Precision * Recall)/(Precision + Recall)

# Overfitting of a tree-

 This a common issue in decision tree ,it happens when tree trie to in dpeth means there are too many branches 
in a tree , it is good to have more branches but some times due to outliers in data tree tries to grow beyond necessary levelsand leaf nodes will also include results due to such ouliers.
And also this makes our model more biased towards test data and it will not be effective when we use it with new dataset.
To avoid it we use Pre-Pruning and Post-Pruning

# Pre-Pruning-

It is a process where we restrict growth of a tree, so when we restrict the growth to certain required level we make sure that 
our tree has no branches due to outliers in data.
To achieve it we set max_depth according to requirement,it is suggested to check accuracy changing the max_dpeth and then select a value of it based on maximum accuracy we get.

Below is the example where we changed max_depth to 6 and we can see slight improvement in accuracy.

```python
dtree3=tree.DecisionTreeClassifier(criterion='entropy',max_depth=6,random_state=0)
dtree3.fit(X_train,y_train)
```

Post-Pruning is the also a usefull method to overcome the overfitting issue , in post-pruning we actually checks effect of our 
pre-pruning ,after the depth we check in post pruning the effect of new nodes , if wesee improvement we continue to increase depth our tree.

# Cross Validation

Crossvalidation is a very effecient way to tune our model, in this the data set is divided into different segments and first one segment is kept seperately(suppose D1) and rest (D2-Dn) are used to train the model now the sperated set is used to test the model. This process is repeated and in next turn second set(D2) is kept and first and others( D1,D3-Dn) are used to train train the model and after that we test the model using second set(D2)and process continues till n times.




What we achieve using Cross validation is that our final model will be more accurate when it will deal with unseen data. It has been seen that even very accurate model does not perform well with unseen data  but using cross -validation we can solve this issue.

Other very important use of cross validation is in case when we have very less number of data to train and test the model.
As we have seen that we use data in segments to train and test model and also we repeat it in that we can train our model more efficiently do deal with new data , even though we had less number of data during development.
